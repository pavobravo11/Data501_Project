```python
import pandas as pd
```

```python
countries = pd.read_excel("European Countries.xlsx")
msci = pd.read_csv("governance/MSCI Ratings Aggregated.csv")
ccpi = pd.read_csv("Environmental/ccpi_2024.csv")
```

```python
model = pd.merge(countries, msci, on = 'Country', how = 'left').merge(ccpi, on = 'Country', how = 'left')
```
