```python
import pandas as pd
```

Data from: https://www.imf.org/en/Publications/WEO/weo-database/2023/October

```python
europe_gdp = pd.read_excel("euro_gdp_data.xlsx")

vars = ['Country', 2023]

nominal_gdp = europe_gdp[europe_gdp['Units'] == "U.S. dollars"][vars]
```

```python
nominal_gdp.reset_index(inplace=True)
```

```python
# Estimates were taken in October 2023, so we will use the exchange rate from that time
# Source, google finance
eur_per_usd = 0.95
```

```python
nominal_gdp['Nominal 2023 GDP (EUR Billions)'] = nominal_gdp[2023]*eur_per_usd
```

```python
nominal_gdp_eur = nominal_gdp[['Country', 'Nominal 2023 GDP (EUR Billions)']]
```

```python
nominal_gdp_eur.tail()
```

```python
# Value in billions of british pounds
jersey_gdp_eur = 5.761*1.15

# Value in Billion USD
liechtenstein_gdp_eur = 7.8267*0.95
```

```python
liechtenstein_gdp_eur
```

```python
# Add the gdp for Jersey and Liechtenstein
jersey = pd.DataFrame({"Country": 'Jersey', "Nominal 2023 GDP (EUR Billions)": jersey_gdp_eur}, index=[0])
liechtenstein = pd.DataFrame({"Country": 'Liechtenstein', "Nominal 2023 GDP (EUR Billions)": liechtenstein_gdp_eur}, index=[0])

nominal_gdp_eur = pd.concat([nominal_gdp_eur, jersey, liechtenstein], ignore_index=True)
```

```python
nominal_gdp_eur.sort_values("Country", ignore_index=True, inplace=True)
```

```python
nominal_gdp_eur
```

```python
nominal_gdp_eur.to_csv("2023_Nominal_GDP_By_Country.csv", index=False)
```
