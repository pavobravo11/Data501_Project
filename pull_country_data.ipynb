```python
from requests import get
import pandas as pd
```

```python
europe_all = get("https://restcountries.com/v3.1/region/europe")
```

```python
europe_countries_set = set()
for i in europe_all:
    europe_countries_set.add(i['name']['common'])
```

```python
europe_countries_df = pd.DataFrame(europe_countries_set, columns=['Name']).sort_values('Name').reset_index(drop=True)
```
